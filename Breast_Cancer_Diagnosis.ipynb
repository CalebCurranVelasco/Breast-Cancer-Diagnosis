```python
import pandas as pd
df = pd.read_csv("wdbc.data.csv")
```

```python
df
```

```python
# malignant 1 benign 0
df["M"] = df["M"].replace("M", 1)
df["M"] = df["M"].replace("B", 0)
```

```python
df
```

```python
del df["index"]
```

```python
df
```

```python
df = df.dropna()
```

```python
# putting data into numpy array
import numpy as np
df_as_np = df.to_numpy()
middle_matrix = df_as_np[:, 2:-1]
X = middle_matrix.reshape(len(df), middle_matrix.shape[1], 1)
y = df_as_np[:, 1]

X.shape, y.shape
```

```python
q80 = round(len(df) * 0.8)
q90 = round(len(df) * 0.9)

x_train, y_train = X[:q80], y[:q80]
x_val, y_val = X[q80:q90], y[q80:q90]
x_test, y_test = X[q90:], y[q90:]

y_train = np.asarray(y_train).astype('float32').reshape((-1,1))
y_val = np.asarray(y_val).astype('float32').reshape((-1,1))
y_test = np.asarray(y_test).astype('float32').reshape((-1,1))
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers

model = Sequential([layers.Input((29,1)),
                    layers.LSTM(128),
                   layers.Dense(32, activation = 'relu'),
                   layers.Dense(32, activation = 'relu'),
                   layers.Dense(1, activation = 'sigmoid')])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

model.fit(x_train, y_train, validation_data=(x_val, y_val), epochs = 100, verbose = 1)
```

```python
test_pred = model.predict(x_test).flatten()
```

```python

def confusion_matrix(act,pred):
    predtrans = ['M' if i > 0.50 else "B" if i < 0.5 else "none" for i in pred]
    actuals = ['M' if i == 1 else "B" for i in act]
    confusion_matrix = pd.crosstab(pd.Series(actuals),
                                  pd.Series(predtrans),
                                  rownames=["Actual"],
                                   colnames=['Predicted'])
    return confusion_matrix
confusion_matrix(y_test, test_pred)
```
